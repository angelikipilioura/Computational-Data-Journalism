````python
# εισάγουμε τις βιβλιοθήκες
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

````python
# κατεβάζουμε τη σελίδα που θέλουμε
response = requests.get('http://www.bbc.co.uk/news')
doc = BeautifulSoup(response.text, 'html.parser')
````

````python
# Ξεκινάμε με μια άδεια λίστα
stories_list = []
stories = doc.find_all('div', { 'class': 'gs-c-promo' })
for story in stories:
    # Φτιάχνουμε ένα λεξικό χωρίς τίποτα
    story_dict = {}
    headline = story.find('h3')
    if headline:
        story_dict['headline'] = headline.text
    link = story.find('a')
    if link:
        story_dict['url'] = link['href']
    summary = story.find('p')
    if summary:
        story_dict['summary'] = summary.text
    # Προσθέτουμε το λεξικό στη λίστα 
    stories_list.append(story_dict)
    
# Δημιουργούμε dataframe από τη λίστα με τα dictionaries
df = pd.DataFrame(stories_list)
````

````python
!pip install newspaper3k
import newspaper
from newspaper import Article
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests


def scrape_site(site_url):
    site = newspaper.build(site_url, memoize_articles=False)

    urlsL = site.article_urls()

    articlesL = []
    for url in urlsL[:10]:
      articleD = {}
      article = Article(url)
      article.download()
      article.parse()
      articleD['url'] = url
      if article.publish_date is not None:
        articleD['date'] = article.publish_date
      else:
        articleD['date'] = np.nan
      if article.title is not None:
        articleD['title'] = article.title
      else:
        articleD['title'] = np.nan
      if article.top_image is not None:
        articleD['top image'] = article.top_image
      else:
        articleD['top image'] = np.nan
      if article.text is not None:
        articleD['text'] = article.text
      else:
        articleD['text'] = np.nan
      if article.authors is not None:
        articleD['authors'] = article.authors
      else:
        articleD['authors'] = np.nan
      if article.images is not None:
        articleD['images'] = article.images
      else:
        articleD['images'] = np.nan
      if article.movies is not None:
        articleD['movies'] = article.movies
      else:
        articleD['movies'] = np.nan
      articlesL.append(articleD)

    return pd.DataFrame(articlesL)

sites = ["https://www.news247.gr/latest/", "http://www.efsyn.gr/"]

for site_url in sites:
    df = scrape_site(site_url)
    print(f"Data from {site_url}:\n")
    print(df)
````

````python
def scrape_site(site_url, num_pages):
    articlesL = []
    
    for page in range(1, num_pages+1):
        page_url = f"{site_url}?page={page}"
        
        site = newspaper.build(page_url, memoize_articles=False)
        urlsL = site.article_urls()
        
        for url in urlsL[:10]:
            articleD = {}
            article = Article(url)
            article.download()
            article.parse()
            
            articleD['url'] = url
            articleD['date'] = article.publish_date if article.publish_date else np.nan
            articleD['title'] = article.title if article.title else np.nan
            articleD['top image'] = article.top_image if article.top_image else np.nan
            articleD['text'] = article.text if article.text else np.nan
            articleD['authors'] = article.authors if article.authors else np.nan
            articleD['images'] = article.images if article.images else np.nan
            articleD['movies'] = article.movies if article.movies else np.nan
            
            articlesL.append(articleD)

    return pd.DataFrame(articlesL)


site_url = "https://www.news247.gr/latest/"
num_pages = 3

df = scrape_site(site_url, num_pages)
print(f"Data from {site_url}:\n")
print(df)
````

````python
def get_news_data(url, headers):
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')

    news_data = []

    articles = soup.find_all('article', {'class': 'teaser__article'})

    for article in articles:
        news_title_tag = article.find('h2', {'class': 'article__title'})
        news_link_tag = article.find('a')
        news_summary_tag = article.find('p', {'class': 'article__leadtext'})
        news_time_tag = article.find('time')

        if news_title_tag and news_link_tag and news_summary_tag and news_time_tag:
            news_title = news_title_tag.get_text(strip=True)
            news_link = news_link_tag['href']
            news_summary = news_summary_tag.get_text(strip=True)
            news_time = news_time_tag['datetime']

            news_data.append({
                'title': news_title,
                'link': news_link,
                'summary': news_summary,
                'datetime': news_time
            })

    return pd.DataFrame(news_data)

headers = {'User-Agent': 'Mozilla/5.0'}
url = 'https://www.news247.gr/latest'

df = get_news_data(url, headers)
print(df)
````

````python
url = 'https://www.news247.gr/latest/?pages=1'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
articles = soup.find_all('article', class_='teaser__article')

data = []

for article in articles:
    title_element = article.find('h2', class_='article__title')
    title = title_element.get_text(strip=True) if title_element else ''

    url_element = article.find('a', href=True)
    url = url_element['href'] if url_element else ''

    date_element = article.find('time')
    date = date_element['datetime'] if date_element else ''

    author_element = article.find('span', class_='article-author')
    author = author_element.get_text(strip=True) if author_element else ''

    category_element = article.find('a', class_='article__section')
    category = category_element.get_text(strip=True) if category_element else ''

    text_element = article.find('p', class_='article__leadtext')
    text = text_element.get_text(strip=True) if text_element else ''

    data.append({
        'title': title,
        'url': url,
        'date': date,
        'author': author,
        'category': category,
        'text': text
    })

df = pd.DataFrame(data)

pd.set_option('display.max_colwidth', 200)

print(df)
````
