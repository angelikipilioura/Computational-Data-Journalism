```python
# This will produce a bar plot that shows the number of movies in each genre.
genre_counts = movies["genre"].value_counts()

plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Number of Movies in Each Genre")
plt.show()

# This will produce a pie chart that shows the distribution of content ratings.
content_rating_counts = movies["content_rating"].value_counts()

plt.pie(content_rating_counts.values, labels=content_rating_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Content Ratings")
plt.show()

# This will produce a histogram that shows the distribution of movie durations.
plt.hist(movies["duration"], bins=20)
plt.xlabel("Duration (min)")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Durations")
plt.show()
```

```python
# Create a histogram of the star ratings
plt.hist(movies["star_rating"], bins=10)
plt.xlabel("Star Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Star Ratings")
plt.show()

# Create a scatter plot of duration vs. star rating
plt.scatter(movies["duration"], movies["star_rating"])
plt.xlabel("Duration (minutes)")
plt.ylabel("Star Rating")
plt.title("Duration vs. Star Rating")
plt.show()

# Create a bar chart of the average star rating by genre
grouped_data = movies.groupby("genre").mean()["star_rating"]
plt.bar(grouped_data.index, grouped_data.values)
plt.xticks(rotation=90)
plt.xlabel("Genre")
plt.ylabel("Average Star Rating")
plt.title("Average Star Rating by Genre")
plt.show()
```
